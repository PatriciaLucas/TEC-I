#Análise de dados




###Boxplots

````python
import pandas as pd
#LEITURA DO ARQUIVO 
dataframe = pd.read_csv('https://raw.githubusercontent.com/PatriciaLucas/TEC-I/main/Datasets/Penguin.csv')
#MOSTRAR OS DADOS
dataframe.head(3)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
sns.set_style("whitegrid")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=[10,4], dpi=100)
g1 = sns.boxplot(x=dataframe['species'], y=dataframe['body_mass_g'], linewidth=0.7, saturation=1,
                 showmeans=True,meanprops={"marker":"o","markerfacecolor":"white", "markeredgecolor":"black","markersize":"5"})

plt.tick_params(labelsize=10)
plt.xlabel("Classes Pinguins", fontsize=10)
plt.ylabel("Body_mass_g", fontsize=10)
pass
````

###Histograma

````python
fig = plt.subplots(figsize=(10,5))
dataframe.groupby("species").body_mass_g.plot(kind='hist', legend=True, title='Massa corporal')
pass
````

###Correlações

````python
dataframe.corr()
````
