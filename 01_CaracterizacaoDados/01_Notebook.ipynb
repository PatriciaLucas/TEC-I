#Geração de dados sintéticos

````python
#PARA PROBLEMAS DE REGRESSÃO
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=80, n_features=1,
                       n_informative=1, noise=20,
                       random_state=0)

#VISUALIZAÇÃO
import pylab as plt
plt.scatter(X, y, marker='o')
````

````python
#PARA PROBLEMAS DE CLASSIFICAÇÃO
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, 
                           n_informative=2, n_clusters_per_class=1, 
                           n_classes=2)

#VISUALIZAÇÃO
import pylab as plt
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)
````

````python
#PROBLEMAS DE CLASSIFICAÇÃO MAIS ESPECÍFICOS
from sklearn.datasets import make_blobs, make_moons, make_circles
X, y = make_blobs(n_samples=500, centers=2, n_features=2, random_state=0, cluster_std=1)

#VISUALIZAÇÃO
import pylab as plt
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)
````

````python
X, y = make_moons(n_samples=500, noise=0.2, random_state=0)

#VISUALIZAÇÃO
import pylab as plt
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)
````

````python
 X, y = make_circles(n_samples=500, noise=0.1, factor=0.5, random_state=1)

 #VISUALIZAÇÃO
import pylab as plt
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y)
````
